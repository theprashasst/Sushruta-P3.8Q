```python
# --- 2. Imports and Configuration ---
print("\n--- [2/5] Importing libraries and configuring model... ---")
import torch
import gradio as gr
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig

# --- IMPORTANT: REPLACE WITH YOUR HUB REPO ID ---
base_model_name="microsoft/Phi-3-mini-4k-instruct"

# ---------------------------------------------

# Your system prompt from training
SYSTEM_PROMPT = """You are Sushruta-P3.8Q, a specialized medical AI assistant from Prashasst's AI Labs. You were created and fine-tuned by Prashasst Dongre to serve as a reliable and accessible educational tool for the public."""

# Quantization configuration for efficient inference
quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16,
)
print("✅ Configuration set.")
```

```python
# --- 3. Load Your Fine-Tuned Model ---
print("\n--- [3/5] Loading your fine-tuned model from the Hub... ---")
# This automatically downloads the base model, quantizes it, and applies your adapters
base_model = AutoModelForCausalLM.from_pretrained(
    base_model_name,
    quantization_config=quantization_config,
    device_map="auto",
    trust_remote_code=True,
)
tokenizer = AutoTokenizer.from_pretrained(base_model_name)
print("✅ Model loaded successfully!")
```

```python
from peft import PeftModel
adapter = "Prashasst/Sushruta-P3.8Q"
model = PeftModel.from_pretrained(base_model, model_id=adapter)
```

```python
chat_history=[]
```

```python
# --- This is the corrected version of your function ---

def generate_response(user_question):
    """
    Takes a new user question and the existing conversation history,
    then returns the model's response.

    Args:
        user_question (str): The new message from the user.
        chat_history (list): A list of message dictionaries, e.g., [{"role": "user", "content": "..."}].
    """

    # 1. Append the new user message to the history
    chat_history.append({"role": "user", "content": user_question})

    # 2. Prepend the system prompt to the full conversation
    #    This ensures the model gets its instructions every time.
    messages_for_model = [{"role": "system", "content": SYSTEM_PROMPT}] + chat_history

    # 3. Use the recommended `apply_chat_template` method
    #    This converts our list of messages into a single, correctly formatted prompt string.
    #    `add_generation_prompt=True` adds the final `<|assistant|>` token to prompt the model to start talking.
    prompt = tokenizer.apply_chat_template(
        messages_for_model,
        tokenize=False,
        add_generation_prompt=True
    )

    # 4. Tokenize the final prompt string
    inputs = tokenizer(prompt, return_tensors="pt", return_attention_mask=False).to(model.device)

    # Keep track of the input length to decode only the new tokens
    input_length = inputs.input_ids.shape[1]

    # 5. Generate the response
    outputs = model.generate(
        **inputs,
        max_new_tokens=1024,
        do_sample=True,
        temperature=0.7,
        top_p=0.95,
        pad_token_id=tokenizer.eos_token_id # Important for generation
    )

    # 6. Slice the output to get only the newly generated tokens
    new_tokens = outputs[0, input_length:]

    # 7. Decode only the new tokens to get the clean assistant response
    assistant_response = tokenizer.decode(new_tokens, skip_special_tokens=True)

    # 8. Append the assistant's response to the history for the next turn
    chat_history.append({"role": "assistant", "content": assistant_response})

    return assistant_response
```

```python
generate_response("Who is Prashasst?")
```

```python
# --- 5. Launch the Gradio App and Create a Public URL ---
print("\n--- [5/5] Launching Gradio... Your public API will be ready shortly. ---")
with gr.Blocks() as demo:
    gr.Markdown("# Sushruta-P3.8Q Demo")
    with gr.Row():
        inp = gr.Textbox(label="Your Question")
        out = gr.Markdown(label="Sushruta's Answer")
    btn = gr.Button("Run")
    btn.click(fn=generate_response, inputs=inp, outputs=out)

# The magic line: `share=True` creates a temporary, public URL for your app.
# This URL is your API endpoint!
demo.launch(share=True, debug=True) # debug=True helps see errors
```
